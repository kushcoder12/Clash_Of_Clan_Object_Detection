```python
from ultralytics import YOLO

# Load a pretrained YOLOv8 model (recommended for training)
model = YOLO("yolov8n.pt")  # yolov8n.pt is a small model

# Train the model
model.train(data="/kaggle/input/clash-of-clan-objectitem-detection/data.yaml", epochs=100, imgsz=640, batch=16)
```

**VALIDATION OF MODEL**

```python
# Evaluate the model performance on the validation set
metrics = model.val(data="/kaggle/input/clash-of-clan-objectitem-detection/data.yaml")
```

**TESTING OF MODEL**

```python
# Test the model performance on the test set
test_results = model('/kaggle/input/clash-of-clan-objectitem-detection/test/images')
print(test_results)
```
